```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import json
from bertopic import BERTopic
import sys
```

```python
df = pd.read_csv('Exportierte Einträge.csv')
```

```python
topic_model = BERTopic(nr_topics = 7, top_n_words=5, verbose=True)
topics, probs = topic_model.fit_transform(df.Title)
```

```python
topic_distr, _ = topic_model.approximate_distribution(df.Title)
```

```python
topics_over_time = topic_model.topics_over_time(df.Title, df['Publication Year'])
```

```python
topics_over_time = topics_over_time[topics_over_time['Topic']!=-1].reset_index(drop=True)
```

```python
topic_model.visualize_topics_over_time(topics_over_time)
```
