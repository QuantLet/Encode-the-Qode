```python
import pandas as pd
import ast 
import tqdm
from tqdm import tqdm
tqdm.pandas()
import re
import pickle
import os
```

```python
DATA_DIR = '../data'
```

```python
with open(os.path.join(DATA_DIR, 'Parsed_Qs_20230126.pkl'), 'rb') as file:
  repos = pickle.load(file)
```

```python
repos['type_script'] = repos['type_script'].str.lower()
```

```python
def parse_metainfo(row):
  #row = re.sub(r"(datetime.date\(.+\))", 
  #             r"'\1'", 
  #             row)
  if row != 'empty':
    #metainfo_row = ast.literal_eval(row)
    metainfo_row = dict((k.lower(), v) for k,v in row.items())
    try:
      meta_q_name = metainfo_row[[qkey for qkey in metainfo_row.keys() if 'quantlet' in qkey][0]]
    except:
      meta_q_name = 'empty'

    meta_desc = metainfo_row['description']
    
    try:
      meta_keywords = metainfo_row[[qkey for qkey in metainfo_row.keys() if 'keyword' in qkey][0]]
    except: 
      meta_keywords = 'empty'
  else:
    meta_q_name = 'empty'
    meta_desc = 'empty'
    meta_keywords = 'empty'
  return meta_q_name, meta_desc, meta_keywords
```

```python
output = repos.metainfo_file.progress_apply(lambda x: parse_metainfo(x))
unzipped_list = list(zip(*output))
```

```python
repos['meta_q_name'] = unzipped_list[0]
repos['meta_desc'] = unzipped_list[1]
repos['meta_keywords'] = unzipped_list[2]
```

```python
with open(os.path.join(DATA_DIR, 'repos_as_pandas_metainfo_20230126.pkl'), 'wb') as file:
  pickle.dump(repos, file)
```
