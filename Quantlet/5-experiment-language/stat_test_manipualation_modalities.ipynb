```python
import pandas as pd

# Example of the Shapiro-Wilk Normality Test
from scipy.stats import shapiro
from scipy.stats import ttest_rel
from scipy.stats import wilcoxon
```

```python
COLUMNS_TEST = ['eval_loss', 'eval_rouge1', 'eval_rouge2', 'eval_rougeL',
        'eval_rougeLsum', 'eval_bleu']
```

```python
# Normal 

results_zs = pd.read_csv('../4-qode2desc/reports/analysis_report_ZS_test_CodeT5_bootstrap/results_bootstrap.csv')
results_zs = results_zs[COLUMNS_TEST].head(34)

results_ft = pd.read_csv('../4-qode2desc/reports/analysis_report_FT_test_CodeT5_bootstrap/results_bootstrap.csv')
results_ft = results_ft[COLUMNS_TEST].head(34)

# Only Words 

py_results_zs = pd.read_csv('../4-qode2desc/reports/analysis_report_ZS_test_CodeT5_bootstrap_only_words/results_bootstrap.csv')
py_results_zs = py_results_zs[COLUMNS_TEST].head(34)

py_results_ft = pd.read_csv('../4-qode2desc/reports/analysis_report_FT_test_CodeT5_bootstrap_only_words/results_bootstrap.csv')
py_results_ft = py_results_ft[COLUMNS_TEST].head(34)


# No Parentheses

r_results_zs = pd.read_csv('../4-qode2desc/reports/analysis_report_ZS_test_CodeT5_bootstrap_removal_operands/results_bootstrap.csv')
r_results_zs = r_results_zs[COLUMNS_TEST].head(34)

r_results_ft = pd.read_csv('../4-qode2desc/reports/analysis_report_FT_test_CodeT5_bootstrap_removal_operands/results_bootstrap.csv')
r_results_ft = r_results_ft[COLUMNS_TEST].head(34)


# Random Order 

m_results_zs = pd.read_csv('../4-qode2desc/reports/analysis_report_ZS_test_CodeT5_bootstrap_random_order/results_bootstrap.csv')
m_results_zs = m_results_zs[COLUMNS_TEST].head(34)

m_results_ft = pd.read_csv('../4-qode2desc/reports/analysis_report_FT_test_CodeT5_bootstrap_random_order/results_bootstrap.csv')
m_results_ft = m_results_ft[COLUMNS_TEST].head(34)
```

```python
# Normal FT
#eval_bleu wilcoxon; rest paired

# Normal ZS
#eval_bleu wilcoxon; rest paired

# Only Words  FT
#eval_bleu wilcoxon; rest paired

# Only Words ZS
#eval_bleu wilcoxon; rest paired

# No Parentheses FT
#eval_bleu - wilcoxon, rest - paired T

# No Parentheses ZS
#eval_rouge1, eval_rougeLsum, eval_bleu - wilcoxon, the rest - paired t


# Random Order FT
# all normal - paired T

# Random Order ZS
#  eval_rouge2 wilcoxon; others normal - paired T;
```

```python
for column in COLUMNS_TEST:
    print(column)
    stat, p = shapiro(results_ft[column])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably Normal')
    else:
        print('Probably not Normal')
```

```python
# Only Words ZS 
for column in COLUMNS_TEST:
    print(column)
    if (column =='eval_bleu'):
        stat, p = wilcoxon(py_results_zs[column], results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
    else:

        stat, p = ttest_rel(py_results_zs[column], results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')

# Only Words FT 
for column in COLUMNS_TEST:
    print(column)
    if (column =='eval_bleu'):
        stat, p = wilcoxon(py_results_ft[column], results_ft[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
    else:

        stat, p = ttest_rel(py_results_ft[column], results_ft[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
```

```python
# No Parentheses ZS 

for column in COLUMNS_TEST:
    print(column)
    if (column =='eval_bleu' or column =='eval_rougeLsum' or column =='eval_rouge1'):
        stat, p = wilcoxon(results_zs[column], r_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
    else:

        stat, p = ttest_rel(results_zs[column], r_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')

# No Parentheses FT

for column in COLUMNS_TEST:
    print(column)
    if (column =='eval_bleu'):
        stat, p = wilcoxon(results_zs[column], r_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
    else:

        stat, p = ttest_rel(results_zs[column], r_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
```

```python
# Random Order ZS
for column in COLUMNS_TEST:
    print(column)
    if (column =='eval_rouge2'):
        stat, p = wilcoxon(results_zs[column], m_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')
    else:

        stat, p = ttest_rel(results_zs[column], m_results_zs[column])
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably the same distribution')
        else:
            print('Probably different distributions')

# Random Order FT
for column in COLUMNS_TEST:
    print(column)
    stat, p = ttest_rel(results_ft[column], m_results_ft[column])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably the same distribution')
    else:
        print('Probably different distributions')
```

```python
# ALL DIFFS ARE NORMAL
for column in ['eval_loss']:
    print(column)
    ow_diff_zs = abs(py_results_zs[column] - results_zs[column])
    ow_diff_ft = abs(py_results_ft[column] - results_ft[column])

    np_diff_zs = abs(r_results_zs[column] - results_zs[column])
    np_diff_ft = abs(r_results_ft[column] - results_ft[column])

    rd_diff_zs = abs(m_results_zs[column] - results_zs[column])
    rd_diff_ft = abs(m_results_ft[column] - results_ft[column])
    
    for i, mode in enumerate([ow_diff_zs, ow_diff_ft, np_diff_zs, np_diff_ft, rd_diff_zs, rd_diff_ft]):
        print(i)
        stat, p = shapiro(mode)
        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            print('Probably Normal')
        else:
            print('Probably not Normal')
```

```python
# ALL DIFFS ARE NORMAL
for column in ['eval_bleu']:
    print(column)
    ow_diff_zs = abs(py_results_zs[column] - results_zs[column])
    ow_diff_ft = abs(py_results_ft[column] - results_ft[column])

    np_diff_zs = abs(r_results_zs[column] - results_zs[column])
    np_diff_ft = abs(r_results_ft[column] - results_ft[column])

    rd_diff_zs = abs(m_results_zs[column] - results_zs[column])
    rd_diff_ft = abs(m_results_ft[column] - results_ft[column])

    stat, p = ttest_rel( ow_diff_zs, np_diff_zs,  alternative='less')
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably not less')
    else:
        print('Probably less')
```
