```python
with open(
    os.path.join("../data", "repos_as_pandas_metainfo_20230126.pkl"), "rb"
) as file:
    repos = pickle.load(file)
repos = repos[repos.type_script == "py"]


def standardize_types(x):
    if isinstance(x, list):
        return ",".join(x)
    else:
        return x


repos.meta_keywords = repos.meta_keywords.apply(standardize_types)
```

```python
del repos["type_script"]
del repos["metainfo_file"]
```

```python
keywords = repos.meta_keywords.copy()
keywords = keywords.str.split(",")
flat_list = [item for sublist in keywords.values for item in sublist]
flat_list = [x.lower().replace(" ", "") for x in flat_list]
freqs = collections.Counter(flat_list)
```

```python
df = pd.DataFrame(data=freqs, index=[0]).T.sort_values(0, ascending=False)
```

```python
df.to_csv("../data/keyword_freqs.csv", index=True)
```
