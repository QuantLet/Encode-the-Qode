```python
import os 
import pickle 
import pandas as pd
import yaml
from datetime import date
```

```python
# PATHS 
PATH_TO_PARSE = '../../data/QuantLet'
FILE_TYPES = ['m', 'py', 'r', 'R', 'M', 'ipynb']
TEST_PATH = '../data/Q_test'

# CONSTANTS
CODE_LOAD = False
IPY_CONVERT = False

# VARIABLE PARAMETERS
TODAY = date.today().strftime("%d%m%Y")
```

```python
if IPY_CONVERT:
    # q_name, folder_name, code_script, type_script, metainfo_file
    language, name = '', ''
    for i, (root, directories, files) in enumerate(os.walk(PATH_TO_PARSE)):
        
        for file in files:
            language = file.split('.')[-1]
            
            if (language == 'ipynb'):
                if not os.path.exists(f"{root}/{file.replace('ipynb', 'py')}"):
                    print(file)
                    os.popen(f"jupyter nbconvert --to python {root}/{file}")
            else:
                continue    
```

```python
Quantlets = []
repos = []
nw = 0
for i, (root, directories, files) in enumerate(os.walk(PATH_TO_PARSE)):

        m_file = [file for file in files if file.lower()=="metainfo.txt"] 
        q = {}
        q['folder_name'] = root
        

        if len(m_file)!=0:
            
            with open(f'{root}/{m_file[0]}', 'r') as meta:
                try:
                    metainfofile = yaml.safe_load(meta)#.readlines()
                except:
                    nw += 1
            q['metainfo_file'] = metainfofile
        else:
            q['metainfo_file'] = 'empty'

        code_files = [file for file in files if file.split('.')[-1] in FILE_TYPES] 
        
        if len(code_files)!=0:

            codes = []
            languages = []
            

            for code_i, c_file in enumerate(code_files):

                if CODE_LOAD:
                        with open(f'{root}/{c_file}', 'rb') as code:
                            sc_file = code.readlines()
                    
                language = c_file.split('.')[-1]

                if code_i == 0:
                    if CODE_LOAD:
                        q['code_script'] = sc_file
                    q['type_script'] = language
                    q['script_name'] = c_file
                    repos.append(q)
                else:
                    extra_q = {}
                    extra_q['metainfo_file'] = q['metainfo_file']
                    extra_q['folder_name'] = q['folder_name']
                    if CODE_LOAD:
                        extra_q['code_script'] = sc_file
                    extra_q['type_script'] = language
                    extra_q['script_name'] = c_file
                    repos.append(extra_q)
 
                #languages.append(language)  
                #    codes.append(sc_file)
                #q['code_script'] = codes
                #q['type_script'] = languages
        else: 
            if CODE_LOAD:
                q['code_script'] = 'empty'
            q['type_script'] = 'empty'

        repos.append(q)
```

```python
repos_df = pd.DataFrame(repos)
if CODE_LOAD:
    repos_df = repos_df[~((repos_df.code_script=='empty')&(repos_df.metainfo_file=='empty')&(repos_df.type_script=='empty'))]
else:
    repos_df = repos_df[~((repos_df.metainfo_file=='empty')&(repos_df.type_script=='empty'))]
repos_df['q_name'] = repos_df.folder_name.str[26:]
repos_df = repos_df.drop_duplicates(['folder_name', 'script_name'])
repos_df
```
